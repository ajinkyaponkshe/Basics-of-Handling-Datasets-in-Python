# Row-wise conditional filtering (single condition) 

# I want rows having year 1967 and country and population columns

```python
# I want rows for year to be 1967 and column country and population
#i want row has only year == 1967 and column=year, country, pop
Year_1967 = gapminder.loc[gapminder['year']== 1967,['year','country','pop']]
Year_1967
```

# Find all entries for the country Afganistan

```python
# find records for Afganistan 
Afganistan= gapminder.loc[gapminder['country']== 'Afghanistan',['Country','Year','pop']]
Afganistan.head()
```

# Row-wise filtering for multiple conditions
 use (case 1) logical operator (case 2)
    use and (&) or (|)
 # records having year as 1967 and pop > 10000000

```python
# use multiple condictions for filtering
# use (case 1) logical operator (case 2)
# records having year as 1967 and pop > 10000000
#  use and (&) or (|)
filter_1=gapminder.loc[(gapminder['year']== 1967) & (gapminder['pop']>1000000),['year','country','pop']]
filter_1.head()
```

```python
# rename columns
gapminder.rename(columns = {'year':'Year','country':'Country'}, inplace = True )
gapminder.head()
```

# Find records where life exp is more than 50 in the year 1967

```python
Life_exp=gapminder.loc[(gapminder['lifeExp']>50) & (gapminder['Year']==1967),['Year','Country','lifeExp']]
Life_exp.head()
```

# Find number of records having life exp more than 50 in the year 1967


```python
#Refer link : https://thispointer.com/pandas-count-rows-in-a-dataframe-all-or-those-only-that-satisfy-a-condition/
#create series for conditions
Countries=gapminder.apply(lambda x: True if x['lifeExp']>50 and x['Year']==1967  else False, axis=1 )
#count records
above_50=len(Countries[Countries==True].index)
above_50
print('Number of records countries having life exp more than 50 are:',above_50)
```

```python
#alternte way
new=gapminder.loc[(gapminder['lifeExp']>50) & (gapminder['Year']==1967)]
len(new)
print('Number of records countries having life exp more than 50 are:',len(new))
```
